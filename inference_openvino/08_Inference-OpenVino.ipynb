```python
# Export

from ultralytics import YOLO

# Load the best-trained model
model = YOLO("/content/drive/My Drive/Colab Notebooks/Final_Dataset/weights/best.pt")

# Export the model to OpenVINO with optimization
model.export(
    format="openvino",
    imgsz=(1024,1024),
    half=False,
    int8=False,
    dynamic=True,
    nms=True,
    batch=1,
    data="/content/drive/My Drive/Colab Notebooks/Final_Dataset/Dataset.yaml"
)
```

```python
# Inference (only BB)

import os
from ultralytics import YOLO

# Load the exported OpenVINO model
ov_model = YOLO("/content/drive/My Drive/Colab Notebooks/Final_Dataset/weights/best_openvino_model", task="detect")

# Path to the test set text file
source = "/content/drive/My Drive/Colab Notebooks/Final_Dataset/autosplit_test.txt"

# Directory for saving inference results
save_dir = "/content/drive/My Drive/Colab Notebooks/Final_Dataset/inference_results"
os.makedirs(save_dir, exist_ok=True)

# Run inference with correct visualization settings
results = ov_model.predict(
    source=source,
    imgsz=(1024, 1024),
    conf=0.43,
    batch=1,
    device="cpu",
    stream=True,
    line_width=1,
    iou=0.5,
    save=True,
    save_txt=True,
    save_conf=True,
    show_boxes=True,
    show_conf=True,
    show_labels=False,
    project=save_dir
)

# Process streamed results
for result in results:
    image_name = os.path.basename(result.path)
    print(f"Processed: {image_name}")

print(f"\n✅ Inference completed! Results saved in '{save_dir}/predict'")
```
