**COMPARE YOLOV8 VARIANT MODELS**

```python
# NANO

import wandb
from wandb.integration.ultralytics import add_wandb_callback
from ultralytics import YOLO

# Initialize a Weights & Biases run
wandb.init(project="my_dissertation", group="Yolov8_Variants", job_type="training", name="nano")

# Load a YOLO model
model = YOLO("yolov8n.pt")

# Add W&B Callback for Ultralytics
add_wandb_callback(model, enable_model_checkpointing=True)

# Fine-Tune the Model
model.train(project="my_dissertation", name="nano", data='/content/drive/My Drive/Colab Notebooks/Final_Dataset/Dataset.yaml', epochs=10, device=0)

# Finalize the W&B Run
wandb.finish()
```
