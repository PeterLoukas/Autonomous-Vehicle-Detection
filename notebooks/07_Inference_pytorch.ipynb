```python
# Inference --No label-conf visualized

import os
from ultralytics import YOLO

# Load the exported model
model = YOLO("/content/drive/My Drive/Colab Notebooks/Final_Dataset/weights/best.pt")

# Path to the test set text file
source = "/content/drive/My Drive/Colab Notebooks/Final_Dataset/autosplit_test.txt"

# Directory for saving inference results
save_dir = "/content/drive/My Drive/Colab Notebooks/Final_Dataset/inference_output"
os.makedirs(save_dir, exist_ok=True)

# Run inference with correct visualization settings
results = model.predict(
    source=source,
    imgsz=(1024, 1024),
    conf=0.43,
    batch=1,
    device='cpu',
    stream=True,
    line_width=1,
    iou=0.5,
    agnostic_nms=True,
    augment=True,
    save=True,
    save_txt=True,
    save_conf=True,
    save_crop=True,
   show_labels=False,
    show_conf=False,
    project=save_dir
)

# Process streamed results
for result in results:
    image_name = os.path.basename(result.path)
    print(f"Processed: {image_name}")

print(f"\n✅ Inference completed! Results saved in '{save_dir}/predict'")
```
